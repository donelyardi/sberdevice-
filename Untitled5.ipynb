## Задание 1

С точки зрения NLP это задача распознавания именованных сущностей или NER. NER решает две связанные подзадачи: обнаружение последовательности токенов, являющейся именованной сущностью и отнесение этой сущности к конкретному классу (например, PER, ORG, LOC)

Для решения этой задачи существует нексколько классических методов. Во-первых, это gazetteers и любые методы типа CRF. Однако на данный момент существуют и более современные методы решения данной задачи: biLSTM, Transformer-based модели

Так же, эту задачу можно решать через LLM. Например: prompting, т.е. методом без дообучения. Формируем промпт: даём текст + инструкцию, просим «вырвать» сущности определённого типа и вернуть их в чётком формате. Еще существует метод с дообучением или fine-tuning. В этом методе мы дообучаем LLM на паре текст - список сущностей

Есть несколько способов оценивать качество модели. Это precision, recall, F1. Их можно рассматривать на уровне токенов или микро и макро усреднения

## Задание 2

```python
import os
import pandas as pd
import re

# Папка с текстами
text_folder = "C:/Users/Даниил/brexit"
# Папка с .out аннотациями
ann_folder = "C:/Users/Даниил/brexi_annotations"

def read_brexit_doc(path: str) -> dict:
    with open(path, "r", encoding="utf-8") as f:
        lines = [line.strip() for line in f.readlines() if line.strip()]
    doc_id = lines[0]
    text = " ".join(lines[4:])
    return {"id": doc_id, "text": text}

docs = []
for fname in os.listdir(text_folder):
    if fname.endswith(".txt"):
        docs.append(read_brexit_doc(os.path.join(text_folder, fname)))

entity_types = ["PER", "ORG", "LOC", "EVT", "PRO"]
rows = []
for doc in docs:
    for ent in entity_types:
        rows.append({
            "document_id": doc["id"],
            "document_text": doc["text"],
            "entity": ent,
            "gold_answer": []
        })

df = pd.DataFrame(rows, columns=["document_id", "document_text", "entity", "gold_answer"])
df["document_id"] = df["document_id"].str.strip()
df["entity"] = df["entity"].str.upper().str.strip()

gold_answers_dicts = {}
for fname in os.listdir(ann_folder):
    if fname.endswith(".out"):
        
        match = re.search(r"_file_(\d+)", fname)
        if match:
            doc_number = match.group(1)
            doc_id = f"ru-{doc_number}"
        else:
            doc_id = fname.replace(".out", "")
        
        annotations = []
        with open(os.path.join(ann_folder, fname), "r", encoding="utf-8") as f:
            for line in f:
                line = line.strip()
                if not line:
                    continue
                parts = line.split("\t")
                if len(parts) >= 3:
                    token, text, entity = parts[:3]
                    annotations.append((text, entity.upper()))
        
        ann_df = pd.DataFrame(annotations, columns=["text", "entity"])
        gold_answers_dicts[doc_id] = {}
        for ent_type in entity_types:
            gold_answers_dicts[doc_id][ent_type] = ann_df.loc[ann_df["entity"]==ent_type, "text"].unique().tolist()

df["gold_answer"] = df.apply(
    lambda row: gold_answers_dicts.get(row["document_id"], {}).get(row["entity"], []),
    axis=1
)

print(df.head())
```

## Задание 3

```python
def make_llm_prompt(row: pd.Series) -> str:
    
    doc_id = row["document_id"]
    entity_type = row["entity"]
    text = row["document_text"]

    entity_map = {
        "PER": "персоны (люди, персонажи)",
        "ORG": "организации (компании, партии, органы власти)",
        "LOC": "географические объекты (страны, города, регионы)",
        "EVT": "события (митинги, выборы, войны, переговоры, саммиты)",
        "PRO": "продукты, документы, технологии, законы"
    }
    entity_description = entity_map.get(entity_type, entity_type)

    prompt = f"""
Выступай в роли системы для извлечения именованных сущностей из новостных текстов на русском языке.

Задача: найти все сущности типа **{entity_type} — {entity_description}** в следующем тексте.
Не добавляй ничего лишнего, не делай обобщений. Сохрани оригинальную форму написания (регистр, склонение и т.д.).

Формат ответа:
Верни результат только в виде JSON-массива строк.
Пример:
["Тереза Мэй", "Борис Джонсон"]

Если сущностей этого типа нет, верни пустой список: [].

Документ ID: {doc_id}
---
{text}
---
Ответ:
""".strip()

    return prompt
```

## Задание 5

Функция ниже вычисляет F1

```python
import numpy as np

def score_fn(gold, pred):

    gold_set = set(gold)
    pred_set = set(pred)
    
    if not gold_set and not pred_set:
        return 1.0  # идеальный случай, если сущностей нет и в gold, и в pred
    
    if not pred_set:
        return 0.0  # ничего не предсказано
    
    tp = len(gold_set & pred_set)  # true positives
    precision = tp / len(pred_set) if pred_set else 0.0
    recall = tp / len(gold_set) if gold_set else 0.0
    
    if precision + recall == 0:
        return 0.0
    
    f1 = 2 * (precision * recall) / (precision + recall)
    return f1
```

```python
def test_score_fn():
    # идеально
    assert score_fn(["A","B"], ["A","B"]) == 1.0
    # пустое предсказание
    assert score_fn(["A","B"], []) == 0.0
    # пустое золото и предсказание
    assert score_fn([], []) == 1.0
    # частично пересекаются
    f1 = score_fn(["A","B"], ["B","C"])
    expected_f1 = 2 * (1/2 * 1/2)/(1/2 + 1/2)  # tp=1, pred_len=2, gold_len=2
    assert np.isclose(f1, expected_f1)
    print("Все тесты пройдены")

test_score_fn()
```

Я выбрал F1, потому что в NER важно находить как можно больше правильных сущностей (recall), но при этом не перепредсказывать (precision).Здесь используется простая реализация, т.к. она корректна для небольших датасетов и легко проверяется через юнит-тесты. С помощью векторизации ускорить можно, но нужно перевести списки сущностей в векторное (бинарное) представление.
